```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x = np.arange(0, 10, 0.1)

# Define fuzzy set
mf = fuzz.trimf(x, [3, 5, 7])

# Plot membership function
plt.plot(x, mf, 'b', linewidth=1.5, label='Triangular')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Membership value')
plt.title('Triangular Membership Function')
plt.legend()
plt.show()
```

```python


# Generate universe variables
x = np.arange(0, 10, 0.1)

# Define fuzzy set
mf = fuzz.gaussmf(x, 5, 2)

# Plot membership function
plt.plot(x, mf, 'r', linewidth=1.5, label='Gaussian')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Membership value')
plt.title('Gaussian Membership Function')
plt.legend()
plt.show()
```
